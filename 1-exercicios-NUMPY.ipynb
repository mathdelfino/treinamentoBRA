### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
array = np.random.rand(10)

print('Matriz original:\n', array, end='\n\n')

# Calula e atribui ao array a raiz aos valores menores ou iguais a 0,5
array[array <= 0.5] = np.sqrt(array[array <= 0.5])

print('Matriz com as raízes dos valores menores ou iguais a 0,5 da matriz original:\n' ,array)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz = np.random.rand(3, 3, 3)

print('Matriz 3D:\n', matriz, end='\n\n')

max_values = np.amax(matriz, axis=1)

print('Maiores valores da linha das matrizes 2D:\n', max_values)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matrix = np.random.rand(5, 5)

print('Matriz gerada:\n', matrix, end='\n\n')

# Média de cada linha
media = np.mean(matrix, axis=1)
print('Média de cada linha: \n', media, end='\n\n')

# Subtrai os valores da linha pela a média de cada linha, utilizando o pivot da media
matrix_media_linha = matrix - media[:, np.newaxis]

print('Matriz subtraída pela média de cada linha:\n', matrix_media_linha)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4, 4)

# Seleciona a diagonal principal
diagonal_principal = np.diag(matriz)

# Seleciona a diagonal secundária
diagonal_secundaria = np.diag(np.fliplr(matriz))

#Calcula a média das diagonais
media_diagonais = np.mean([diagonal_principal, diagonal_secundaria])

#Calcula o desvio padrão das diagonais
dp_diagonais = np.std([diagonal_principal, diagonal_secundaria])

print('Matriz gerada:\n', matriz, end='\n\n')
print('Diagonal principal:\n', diagonal_principal, end='\n\n')
print('Diagonal secundária:\n', diagonal_secundaria, end='\n\n')
print('Média das diagonais:', media_diagonais, end='\n\n')
print('Desvio padrão das diagonais:', dp_diagonais)
```
